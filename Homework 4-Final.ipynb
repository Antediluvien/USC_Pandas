# Top Spenders

```python
#Find the total amount spent
SN_Price_group = df.groupby(['SN'])['Price'].sum().to_frame()
SN_Price_group = SN_Price_group.sort_values(by = "Price", ascending=False)
SN_Price_group.reset_index(level=0, inplace=True)
SN_Price_group.columns = ['Player Name', 'Total Spent']
SN_Price_group.head()
```

```python
#Find the amount of Purchases made
Purchase_amount_Player = df['SN'].value_counts().to_frame()
Purchase_amount_Player.reset_index(level=0, inplace=True)
Purchase_amount_Player.columns = ['Player Name', 'Purchase Count']
Purchase_amount_Player.head()
```

```python
#Combine both dataframe of purchase count and money spent
Top_Spender_df = pd.merge(SN_Price_group, Purchase_amount_Player, on=['Player Name'])
Top_Spender_df['Average Purchase Price'] = Top_Spender_df['Total Spent'] / Top_Spender_df['Purchase Count']
Top_Spender_df['Total Spent'] = Top_Spender_df['Total Spent'].map('${:,.2f}'.format)
Top_Spender_df['Average Purchase Price'] = Top_Spender_df['Average Purchase Price'].map('${:,.2f}'.format)
Top_Spender_df.head()
```


# Most Popular Items

```python
df_items = df[['Item ID','Item Name','Price']]
df_items['Purchase Count'] = df.groupby(['Item ID'])['Price'].transform('count')
df_items = df_items.sort_values('Purchase Count',ascending=False)
pd.options.mode.chained_assignment = None
df_items = df_items.drop_duplicates()
df_items['Total Revenue'] = df_items.Price * df_items['Purchase Count']
df_items['Price'] = df_items['Price'].map('${:,.2f}'.format)
df_items['Total Revenue'] = df_items['Total Revenue'].map('${:,.2f}'.format)
df_items.head()
```
